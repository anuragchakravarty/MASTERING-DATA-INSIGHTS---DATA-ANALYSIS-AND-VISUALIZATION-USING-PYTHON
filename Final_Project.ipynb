### Calculating Mean, Median, Mode, Standard deviation and Percentile ranks

```python
# Extracting only the marks for performing further calculations

data = dataset[:, 1].astype(float)
```

```python
# Mean

mean = np.mean(data)
print(f"Mean of the students marks is : {mean:.2f}")
```

```python
# Median

median = np.median(data)
print(f"Median of the students marks is : {median:.2f}")
```

```python
# Mode

from scipy import stats                 #for performing calculation of mode

mode = stats.mode(data)
print(f"Mode is : {mode}")
```

```python
# Standard deviation

standard_deviation = np.std(data)
print(f"Standard deviation of the data is : {standard_deviation:.2f}")
```

```python
# Percentile ranks 

data_sorted = np.sort(data)

# Percentile ranks calculation
percentiles = [10, 25, 50, 75, 90]  

for percentile in percentiles:
    percentile_value = np.percentile(data_sorted, percentile)
    print(f"Percentile ({percentile} %): {percentile_value:.2f}")
```

### Using pandas to create and manipulate DataFrames containing student scores and demographics.

```python
import pandas as pd

df = pd.DataFrame(dataset, columns=['Student Name', 'Student Scores','Age','Gender'])
df
```

### Using Matplotlib to analyzing performance trends involves examining the data to identify patterns and insights regarding student scores.

```python
import matplotlib.pyplot as plt
```

```python
# Extracting names and marks column

names = df['Student Name'].tolist()
marks = df['Student Scores'].tolist()
age   = df['Age'].tolist()
gender= df['Gender'].tolist()
```

```python
# Analyzing performance trends (Marks vs. Student column).

plt.figure(figsize=(12, 6))
plt.title("Distribution of Student's Marks")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.bar(names, marks, color='skyblue')
plt.xticks(rotation=90)
plt.grid(axis='y')

# Display the Bar plot
plt.show()
```

### Using Seaborn to analyzing correlations

```python
import seaborn as sns
```

```python
df['Marks'] = df['Student Scores'].astype(float)
df['Age'] = df['Age'].astype(int)
```

```python
numeric_df = df[['Student Scores', 'Age']]
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
